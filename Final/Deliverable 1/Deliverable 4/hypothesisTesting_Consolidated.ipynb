## Comparing percentage of fully vaccinated population

```python
from scipy.stats import ttest_ind

city = ['Chelsea', 'Newton', 'Revere', 'Wellesley','Springfield', 'Everett']
race = ['Black','Hispanic','White','Asian']

def t_test(df1, df2):
	stat,p = ttest_ind(df1, df2 ,equal_var = False)
	print('stat=%.3f, p=%.8f' % (stat, p))
	if p > 0.05:
		print('Probably the same distribution')
	else:
		print('Probably different distributions')

	return p


def hypothesis_test(df):
	for i in range(len(city)-1):
		for j in range(i+1, len(city)):
			df1 = df[df['city'] == city[i]]
			df2 = df[df['city'] == city[j]]
			print(city[i], 'vs.', city[j])
			for a in range(len(race)):
				df_1 = df1[df1['race'] == race[a]]
				df_2 = df2[df2['race'] == race[a]]
				print(race[a])
				t_test(df_1['percentage_fully_vaccinated'], df_2['percentage_fully_vaccinated'])

			print('\n')

hypothesis_test(df)
```

# Rate of Change

```python
from scipy.stats import ttest_ind

city = ['Chelsea', 'Newton', 'Revere', 'Wellesley','Springfield', 'Everett']
race = ['Black','Hispanic','White','Asian']

def t_test(df1, df2):
	stat,p = ttest_ind(df1, df2 ,equal_var = False)
	print('stat=%.3f, p=%.8f' % (stat, p))
	if p > 0.05:
		print('Probably the same distribution')
	else:
		print('Probably different distributions')

	return p


def hypothesis_test(df):
	for i in range(len(city)-1):
		for j in range(i+1, len(city)):
			df1 = df[df['city'] == city[i]]
			df2 = df[df['city'] == city[j]]
			print(city[i], 'vs.', city[j])
			for a in range(len(race)):
				df_1 = df1[df1['race'] == race[a]]
				df_2 = df2[df2['race'] == race[a]]
				diff1 = df_1['percentage_fully_vaccinated'].diff().dropna()
				diff2 = df_2['percentage_fully_vaccinated'].diff().dropna()
				print(race[a])
				t_test(diff1, diff2)

			print('\n')

hypothesis_test(df)
```

### Summary


We attempted to figure out if there is a significant difference in rate of change of fully vaccinated population. We took the difference of percentage of the fully vaccinated population between two consecutive times and then used hypothesis testing to examine how similar they are.

When comparing Chelsea and other cities, surprisingly we found that the rates of change for hispanic, black, asisn and white population are all statistically significant because their p-value are greater than the significance level (0.05).
