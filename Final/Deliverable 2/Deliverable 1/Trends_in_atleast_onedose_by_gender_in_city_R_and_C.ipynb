```python
import pandas as pd

df = pd.read_csv ('Revere_sex.csv')
df.head()

#df['date']=df['date'].str[:7]
#df['date']
```

```python
import matplotlib.pyplot as plt
import pandas


gender_set = set(df['gender'])

plt.figure()
for gender in gender_set:
     selected_data = df.loc[df['gender'] == gender]
     plt.plot(selected_data['date'], selected_data['atleast_onedose'], label=gender)

axes = plt.axes()
x_values = axes.get_xticks()
x_len = len(x_values)
new_x = [x_values[i] for i in [0, x_len // 3,x_len*2 // 3, -1]]
axes.set_xticks(new_x)
plt.title('Trends in atlear_onedose by gender in Revere')

plt.legend()
plt.show()
```
