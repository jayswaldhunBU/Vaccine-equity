```python
df_chelsea = df[df['city'] =='Chelsea']
df_newton = df[df['city'] =='Newton']
df_revere = df[df['city'] =='Revere']
df_wellesley = df[df['city'] == 'Wellesley']
```

Hypothesis Testing Between Newton and Chelsea:

Hypothesis1:
H0: The mean of fully vaccinated for black people in Chelsea is the same as that in Newton.
Ha: The mean of fully vaccinated for black people in Chelsea is different from that in Newton.

```python
# Example of the Student's t-test
from scipy.stats import ttest_ind
# Example of the Analysis of Variance Test
from scipy.stats import f_oneway

df_chelsea_black = df_chelsea[df_chelsea['race'] == 'Black']
df_newton_black = df_newton[df_newton['race'] == 'Black']

def t_test(df1, df2):
	stat,p = ttest_ind(df1, df2 ,equal_var = False)
	print('stat=%.3f, p=%.8f' % (stat, p))
	if p > 0.05:
		print('Probably the same distribution')
	else:
		print('Probably different distributions')

	return p

t_test(df_newton_black['fully_vaccinated'], df_chelsea_black['fully_vaccinated'])
```

```python
city = ['Chelsea', 'Newton', 'Revere', 'Wellesley']
race = ['Black','Hispanic','White']
def hypothesis_test(df):
	for i in range(len(city)-1):
		for j in range(i+1, len(city)):
			df1 = df[df['city'] == city[i]]
			df2 = df[df['city'] == city[j]]
			print(city[i], 'vs.', city[j])
			for a in range(len(race)):
				df_1 = df1[df1['race'] == race[a]]
				df_2 = df2[df2['race'] == race[a]]
				print(race[a])
				t_test(df_1['fully_vaccinated'], df_2['fully_vaccinated'])

			print('\n')

hypothesis_test(df)
```
